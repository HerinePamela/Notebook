```python
def graph(y):
    sns.boxplot(x="Species", y=y, data=sample_ir)
  
plt.figure(figsize=(12,12))
      
# Adding the subplot at the specified
# grid position
plt.subplot(221)
graph('SepalLengthCm')
  
plt.subplot(222)
graph('SepalWidthCm')
  
plt.subplot(223)
graph('PetalLengthCm')
  
plt.subplot(224)
graph('PetalWidthCm')
  
plt.show()
```

```python
fig, axes = plt.subplots(1,  figsize=(9,9))
sns.boxplot(x='SepalWidthCm', data=sample_ir)
```

```python
Q1 = np.percentile(sample_ir['SepalWidthCm'], 25,
                interpolation = 'midpoint')
  
Q3 = np.percentile(sample_ir['SepalWidthCm'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1
  
print("Old Shape: ", sample_ir.shape)
  
# Upper bound
upper = np.where(sample_ir['SepalWidthCm'] >= (Q3+1.5*IQR))
  
# Lower bound
lower = np.where(sample_ir['SepalWidthCm'] <= (Q1-1.5*IQR))
  
# Removing the Outliers
sample_ir.drop(upper[0], inplace = True)
sample_ir.drop(lower[0], inplace = True)
  
print("New Shape: ", sample_ir.shape)
  
sns.boxplot(x='SepalWidthCm', data=sample_ir)
```
