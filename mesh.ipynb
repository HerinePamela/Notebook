```python
!pip install pygimli
import pygimli as pg
import pygimli.meshtools as mt
```

```python
left = -30
right = 30
depth = 25
```

```python
 world = mt.createWorld(start = [left, 0],
               end = [right, -depth],
               layers=[-5])
 pg.show(world);
```

```python
line = mt.createLine(start = [left, -20], end = [right, -15])
pg.show(line);
```

```python
geometry = world + line
pg.show(geometry)
```

```python
body = mt.createPolygon([[-8, -12], [0, -7], [8,-13]],
                        isClosed = True, marker = 3,
                        addNodes =5, interpolate = 'spline')
pg.show(body, showNodes = True);

geometry += body
pg.show(geometry);
```

```python
print(geometry)
```

```python
pg.show(geometry, boundaryMarkers = True)
```

```python
 mesh = mt.createMesh(geometry, quality = 33.5, area = 1, smooth = True)
 pg.show(mesh);
```

```python
mesh.save('mesh.bms')
```

```python
mt.exportPLC(geometry, 'geometry')
```

MESH MODIFICATION

```python
translate_mesh = pg.Mesh(mesh)
translate_mesh.translate([500, 25])
pg.show(translate_mesh)
```

```python
scaled_mesh = pg.Mesh(mesh)
scaled_mesh.scale([1,2])
pg.show(scaled_mesh);
```

```python
import numpy as np
rotated_mesh = pg.Mesh(mesh)
rotated_mesh.rotate([0, 0,  np.deg2rad(-20)])
pg.show(rotated_mesh)
```

```python
# 3D EXTRUSION
extruded_mesh = mt.extrudeMesh(mesh, np.linspace(0, 20, 5))
extruded_mesh.rotate([np.pi/2,0,0])
pg.show(extruded_mesh, extruded_mesh.cellMarkers());
```

```python
cube = mt.createCube(size = [5, 5, 5])
cylinder = mt.createCylinder(height = 5, pos = [0,0,5], marker = 2, nSegments = 20)
geometry3D = cube + cylinder
mesh3D = mt.createMesh(geometry3D, area = 0.1)
mesh3D.rotate([0, -np.pi/8, 0.0])
pg.show(mesh3D, mesh3D.cellMarkers());
```
